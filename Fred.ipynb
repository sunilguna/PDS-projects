```python
%matplotlib inline
import matplotlib.ticker as plticker
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt

#Check if the string is a floating point number
def is_float(string):
    try:
        float(string)
        return True
    except ValueError:
        return False
    
def fred(series,frequency,units, start_date,end_date):
    result = []
#Selection parameters
    selection = { 'api_key' : '770130e3e272d1085975a8e920dd6aa1',
                  'file_type' : 'json',
                  'series_id' : series, 
                  'frequency' : frequency,
                  'units' : units,
                  'observation_start' : start_date,
                  'observation_end' : end_date }
    
    r = requests.get('http://api.stlouisfed.org/fred/series/observations?', params=selection)
    result = r.json()
    obs = dict()
#Storing the observations
    for i,x in enumerate(result["observations"]):
        key = x["date"]
        if is_float(x["value"]):
            obs[key] = float(x["value"])
#Convert the observations into a series
    result = pd.Series(obs)
    return result

#Yearly data plot
freq = "q"
units = "pch"
date_start = "2000-01-01"
date_end = "2013-11-30"

obs_1 = fred("SP500", freq, units, date_start, date_end)
obs_2 = fred("WILLWWW", freq, units, date_start, date_end)

obs_1.plot(color = 'g', label = 'S&P')
obs_2.plot(color = 'r', label = 'Internet Index')
plt.xlabel('Time')
plt.ylabel('Change')
plt.xticks(rotation = 90)
plt.legend()

plt.figure()

#Daily Plot for data collected
freq = "d"
units = "pch"
date_start = "2013-09-11"
date_end = "2013-12-06"

obs_1 = fred("SP500", freq, units, date_start, date_end)
obs_2 = fred("WILLWWW", freq, units, date_start, date_end)

obs_1.plot(color = 'g', label = 'S&P')
obs_2.plot(color = 'r', label = 'Internet Index')
plt.xlabel('Time')
plt.ylabel('Change')
plt.xticks(rotation = 90)
plt.legend()
```
